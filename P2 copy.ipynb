## Create paths without backtracking

```python
def removeBacktracking(path):
    st = []
    for p in path:
        if p == '<':
            if len(st) == 0:
                exit(1)
            st.pop()
        else:
            st.append(p)
    return st
```

```python
def compare_edge_reusage(df_1, df_2):
    # Since dictionaries are sorted we only fetch first element and determine max
    row_1 = df_1.iloc[0]['Count']
    row_2 = df_2.iloc[0]['Count']
    max_compare = min(row_1,row_2)
    df1_counts = []
    df2_counts = []
    for i in range(2,max_compare):
        df_1_filtered_count = df_1[df_1['Count'] > i].size
        df_2_filtered_count = df_2[df_2['Count'] > i].size
        
        df1_counts.append(df_1_filtered_count)
        df2_counts.append(df_2_filtered_count)
    return df1_counts, df2_counts
```

### Determine bag of starting articles for each highway

```python
def find_start_end_bags_using_highway(highway, df_finished):
    ## returns filter
    length = len(highway)
    start_words=[]
    end_words = []
    for path in df_finished:
        if len(path)-2 < length:
            continue
        for i in range(1,len(path)-length-1):
            tup = tuple(path[i:i+length])
            if tup == highway:
                start_words.append(path[i-1])
                end_words.append(path[i+length+1])
    return start_words, end_words
```

## Research Question 2

In this part, we try to find out how different player complete the game in different ways, and catogrize all the paths into different groups. We then focus on those paths that are differnt from the majority.

In general, the basis of the classification is the change in distance to the destination of each step. Specifically, when moving from one page to another through the link, the player may get closer to the destination if he or she chooses the correct link, which will decrease the distance to the destination by 1. On the contrary, the player may also get further to the destination. We then count in the whole path, how many edges have made the distance decreases, how many edges have not change the distance, and how many edges have increase the distance by 1, 2 or more. We can then get a table like this:

| Distance Change | inf (move to pages that can't reach the destination) | -1 | 0 | 1 | 2 | 3 |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| count | 0 | 4 | 2 | 2 | 0 | 0 |

With such information, we can characterize each path with a vector of distance change count. For example, the path of the above table can be characterize with the vector [0, 4, 2, 2, 0, 0]. If we normalize the vector, we can get [0, 0.5, 0.25, 0.25, 0, 0].

With each path characterized with a vector, we can apply data analyzing methods like clustering on the vector space, and get different groups of paths.

```python
# load shortest path distance matrix
spd_matrix = []
for line in open('./data/wikispeedia_paths-and-graph/shortest-path-distance-matrix.txt', "r"):
    if ((line[0] == '#') or (line[0].isspace())):
        continue
    spd_matrix.append(line)
```

```python
df_articles['id'] = df_articles.index

df_articles_reindex = df_articles.set_index('article')
df_articles_reindex.head()

def computeEdgeDir(path, target = None, normalize = False, shortest_edge_list = False):
    dir_distribute = np.zeros(6)
    st = []
    shortest = []
    if (target == None):
        target = path[-1]
    tid = df_articles_reindex['id'][target]
    for p in path:
        if p == '<':
            st.pop()
            continue
        else:
            st.append(p)
        if (len(st) <= 1):
            continue
        u = st[-2]
        v = st[-1]
        uid = df_articles_reindex['id'][u]
        vid = df_articles_reindex['id'][v]
        if (spd_matrix[vid][tid] == '_'):
            dir_distribute[0] += 1
        elif (spd_matrix[uid][tid] == '_'):
            print(path)
            if (vid == tid):
                dir_distribute[1] += 1
        else:
            dir_distribute[int(spd_matrix[vid][tid]) - int(spd_matrix[uid][tid]) + 2] += 1
            if (int(spd_matrix[vid][tid]) - int(spd_matrix[uid][tid]) + 2 == 1):
                shortest.append((u, v))
    if (shortest_edge_list):
        return shortest
    if (normalize):
        if (dir_distribute.sum() == 0):
            print(path)
        else:
            dir_distribute /= dir_distribute.sum()
    return dir_distribute

df_finished_paths['edge_direction_distribution'] = df_finished_paths['path_list_without_backtracking'].apply(lambda x: computeEdgeDir(x))
df_finished_paths['normalized_edge_direction_distribution'] = df_finished_paths['path_list_without_backtracking'].apply(lambda x: computeEdgeDir(x, normalize = True))
df_finished_paths['shortest edges'] = df_finished_paths['path_list_without_backtracking'].apply(lambda x: computeEdgeDir(x, shortest_edge_list = True))

#df_finished_paths['edge_direction_distribution']
```

```python
# get the adjacency matrix from the shortest path distance matrix
spd_matrix_num = np.ones([len(spd_matrix), len(spd_matrix)])
spd_matrix_np = np.array(list(map(lambda x: np.array([*x][:-1]), spd_matrix)))
spd_matrix_num[spd_matrix_np != '1'] = 0
```

```python
# function for computing the shortest path with dijkstra
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import dijkstra
edges = csr_matrix(spd_matrix_num)

def shortest_path(x):
    s = x[0]
    t = x[-1]
    sid = df_articles_reindex['id'][s]
    tid = df_articles_reindex['id'][t]
    sp_dist, sp_pred = dijkstra(csgraph = edges, directed = True, indices = sid, return_predecessors = True)
    path = [t]
    now = tid
    while (now != sid):
        now = sp_pred[now]
        path.append(df_articles.iloc[now]['article'])
    path.reverse()
    return path

df_most_detour['shortest_path'] = df_most_detour['path_list_without_backtracking'].apply(lambda x: shortest_path(x))
```

```python
# function for retriving other players' path
def other_player_shortest(x):
    s = x[0]
    t = x[-1]
    other_shortest = df_finished_paths[(df_finished_paths["start"] == s) & (df_finished_paths["destination"] == t)].sort_values(by = 'path_length_without_backtracking').iloc[0]['path_list_without_backtracking']
    if (len(other_shortest) == len(x)):
        return None
    return other_shortest

df_most_detour['other_player_shortest'] = df_most_detour['path_list_without_backtracking'].apply(lambda x: other_player_shortest(x))

df_most_detour[['path_list_without_backtracking', 'shortest_path', 'other_player_shortest']].head(50)
```

3. Lingering paths. These paths quickly reach the pages that are semantically close to the destination, but, unfortunately, they are just semantically related (without a hyperlink). In the following example, we can see that the first path goes to 'football' quickly, but failed to find a way from 'football' to 'sport'. Finally it takes detours to find the correct way. In the second path, it goes through Asian countries, regions and cultures, but failed to find a way to the page 'Asia' itself.

```python
df_most_detour[['path_list_without_backtracking', 'shortest_path', 'other_player_shortest']].iloc[[18, 0]]
```

In fact, the boundaries of the three types are not absolute, a detoured path can be both sequential retrival path and also lingering path.
